# **Import Trained Model**

```python
import gensim.downloader as api
model=api.load("word2vec-google-news-300")
```

# **Example Of Word As Vector**

```python
word_vectors=model
```

```python
print(word_vectors['zebra'])
```

```python
print(word_vectors.most_similar(positive=['King','Woman'],negative=['Man'],topn=10))
```

```python
print(word_vectors['vector'].shape)
```

# SimilaritY Between A Pair Of **Words**

```python
print(word_vectors.similarity('Husband','Wife'))
print(word_vectors.similarity('Father','Son'))
print(word_vectors.similarity('Cat','Kitten'))
print(word_vectors.similarity('Rich','Money'))
print(word_vectors.similarity('Brother','Sister'))
print(word_vectors.similarity('Knowledge','Job'))
print(word_vectors.similarity('Job','Money'))
print(word_vectors.similarity('Intelligence','Money'))
print(word_vectors.similarity('Uncle','Aunt'))
```

```python
print(word_vectors.most_similar('Money',topn=3))
```

# **Create Token Embeddings**

```python
import torch
```

```python
input_ids=torch.tensor([2,3,5,1])
```

```python
vocab_size=6
output_dim=3
```

```python
torch.manual_seed(123)
embedding_layer=torch.nn.Embedding(vocab_size,output_dim)
```

```python
print(embedding_layer.weight)
```

```python
print(embedding_layer.weight[3])
```

```python
print(embedding_layer(torch.tensor([3])))
```

```python
print(embedding_layer(input_ids))
```
