```python
def binary_cross_entropy(y_pred,y):
  return -(y*torch.log(y_pred)+(1-y)*torch.log(1-y_pred))
```

```python
loss=binary_cross_entropy(y_pred,y)
loss
```

```python
loss.backward()
w.grad
```
